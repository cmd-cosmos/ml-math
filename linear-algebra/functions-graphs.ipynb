```python
# geometric intuition for functions
'''
Notes:
- all odd functions return -1 for a +1 value, i.e they return -ve for -ve
- all even functions return a +ve value for -ve inputs except for square root(imaginary out.)
- odd functions do have symmetry across origin but not across y-axis
- y-axis is the range
- x-axis is the domain of the function
'''
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
# Constant functions --> f(x) = b
# all values of x return a contant value b
# domain ===> all Real nums
# range ===> the constant value supplied
# even function ---> symmetric along the y-axis

def constant_fn(x):
    return 10 # b = 10

x = [i for i in range(0, 10)]
y = [constant_fn(i) for i in x] 

plt.plot(x, y)
```

```python
# Linear functions ---> f(x) = x 
# linear growth of values
# odd function ---> assymetric along the y-axis, always increasing
# domain ===> all Real nums
# range ===> all real nums
# every odd function will have a -1 return for a +1 input

def linear_fn(x):
    return x

x = [i for i in range(-10,11)]
y = [linear_fn(j) for j in x]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.plot(x,y)
```

```python
# Quadratic function ---> f(x) = x^2
# even function ---> symmetric across the y-axis looks like a parabola
# domain ===> all real numbers
# range ===> all +ve real numbers


def quadratic_fn(x):
    return x**2

x = [i for i in range(-20,21)]
y = [quadratic_fn(j) for j in x]

plt.axhline(y=0, color='red',linestyle='dashed')
plt.axvline(x=0, color='red',linestyle='dashed')
plt.plot(x,y)
```

```python
# Cubic function ---> f(x) = x^3
# odd function and always increasing
# domain ===> all Real nums
# range ===> all real nums

def cubic_fn(x):
    return x**3

x = [i for i in range(-20,21)]
y = [cubic_fn(j) for j in x]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.plot(x,y)
```

```python
# Modulus/Absolute function ---> f(x) = |x|
# even function
# domain ===> all real numbers
# range ===> all +ve real numbers

def absolute_fn(x):
    return abs(x)

x = [i for i in range(-20,21)]
y = [absolute_fn(j) for j in x]

plt.plot(x,y)
plt.axhline(y=0, color='red',linestyle='dashed')
plt.axvline(x=0,color='red',linestyle='dashed')
```

```python
# Cube root function ---> f(x) = x^(1/3)
# odd function and always increasing
# domain ===> all real nums
# range ===> all real nums

def cuberoot_fn(x):
    return (abs(x))**(1/3)

x = [i for i in range(-1000,1000)]
y = [-1*cuberoot_fn(j) if j<0 else cuberoot_fn(j) for j in x]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.plot(x,y)
```

```python
# Square Root function ---> f(x) = sqrt(x)
# neither even nor odd as no reflection along the y-axis so no comments on symmetry
def sqrt_fn(x):
    return math.sqrt(x)

x = [i for i in range(-1000,1000)]
y = [sqrt_fn(j) if j > 0 else None for j in x]

plt.plot(x,y)
plt.axhline(y=0, color='green', linestyle='dashed')
plt.axvline(x=0, color='green', linestyle='dashed')
```

```python
# Reciprocal function ---> f(x) = 1/x
# odd function and always decreasing
# vertical assymptotes can never be touched
# horizontal assmptotes can be crossed within certain ranges but that gets nullified as we approach infinity
# domain ===> all Real nums except 0
# range ===> all real nums except 0

def reciprocal_fn(x):
    return 1/x

x = [i for i in range(-20,21)]
y = [reciprocal_fn(j) if j!=0 else None for j in x]

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.plot(x,y)
# plotting assymptotes
plt.axhline(y=0,color='red', linestyle='dashed')
plt.axvline(x=0,color='red',linestyle='dashed')
```

```python
# Logarithmic function
# neither even nor odd, domain does not include -ve values
# domain ===> all +ve real numbers
# range ===> all real numbers


def log2_fn(x):
    return math.log(x,2) # base 2 log

def ln_fn(x):
    return math.log(x) # natural log
    
def log10_fn(x):
    return math.log(x,10) # base 10 log
    
x = [i for i in range(-100,100)]
y = [log2_fn(j) if j>0 else None for j in x]
y2 = [ln_fn(j) if j>0 else None for j in x]
y3 = [log10_fn(j) if j>0 else None for j in x]

plt.plot(x,y, label='log base 2')
plt.plot(x,y2, label='log base e')
plt.plot(x,y3, label='log base 10')
plt.legend()
plt.axhline(y=1, color='red', linestyle='dashed')
```

```python
# Exponential function ---> f(x) = e^x
# neiher even nor odd in nature
# domain ==> all real numbers
# range  ==>all +ve real numbers

def exp_fn(x):
    return np.exp(x)

x = [i for i in range(-5,5)]
y = [exp_fn(j) for j in x]

plt.plot(x,y)
plt.axhline(y=0, color='red', linestyle='dashed')
plt.axvline(x=0, color='red', linestyle='dashed')
```
