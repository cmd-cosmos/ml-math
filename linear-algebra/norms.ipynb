### l2 norm ----> Euclidean Distance ----> distance between 2 pts x and y via ||x-y||2

 

```python
import numpy as np

p1 = np.array([1,2,3])
p2 = np.array([10,20,30])

res1 = np.linalg.norm(p1-p2)
print("Euclidean Distance method 1: ", res1)

res2 = np.sqrt(np.sum((p1-p2)**2))
print("Euclidean Distance method 2: ", res2)

res3 = np.sqrt(np.dot(p1 - p2, p1 - p2))
print("Euclidean Distance method 3: ", res3)
```

### l1 norm:
The  ℓ1 norm of a vector x is defined as the sum of the absolute values of its entries.




```python
arr = np.array([1,-2,3,-4,5,-6])
l1_norm = np.linalg.norm(arr, 1)

print("l1 norm: ", l1_norm)
```

### infinite norm of a vector/matrix:
The  ℓ∞ norm of a vector x is equal to its largest entry in terms of absolute value.




```python
arr2 = np.arange(12).reshape(3,4) # create a 3x4 matrix with entries in range [1,11]
l_inf = np.linalg.norm(arr2, np.inf)

# essentially calculated using the absolute sum of the rows of the matrix.
print("matrix: \n", arr2)
print("infinite norm of matrix: ", l_inf)
```
