```python
'''Variance and standard deviation using python'''

import statistics as stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
nums = [1,1,1,2,2,2,2,2,2,7,7,7,9,9,9000,10,1000000]

print('mean: ', stats.mean(nums))           # distribution will be skewed to the right
print('median: ', stats.median(nums))       # median of the distribution
print('mode: ', stats.mode(nums))           # mode of the distribution

print('std deviation: ', stats.stdev(nums)) # extremely high standard deviation
print('variance: ', stats.stdev(nums))      # same value for variance
```

```python
sns.histplot(nums,bins=5, kde=True)
```
