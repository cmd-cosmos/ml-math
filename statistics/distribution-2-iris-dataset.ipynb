```python
'''
Basic statistics on the iris dataset
'''
import statistics as stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# measures of central tendancy for the sepal length column

print('mean = ',np.mean(df['sepal_length']))
print('median = ',np.median(df['sepal_length']))
print('mode = ',stat.mode(df['sepal_length']))
```

```python
sns.boxplot(df['sepal_length'],orient='h')
```

```python
sns.histplot(df['sepal_length'],kde=True)

# not a gaussian as spread is visible near mean values.
```

```python
# measures of central tendancy for the sepal width column

print('mean = ',np.mean(df['sepal_width']))
print('median = ',np.median(df['sepal_width']))
print('mode = ',stat.mode(df['sepal_width']))
```

```python
sns.boxplot(df['sepal_width'],orient='h')
```

```python
sns.histplot(df['sepal_width'],kde=True)

# looks like a gaussian distribution
```

```python
sns.countplot(df, x='species')
```
